```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 평균 필터 함수 생성 
# 현재 측정값 z, α를 계산하기 위한 데이터 개수 k, 직전 평균값인 pre_avg를 입력으로 받아 
# 현재 측정값까지의 평균 반환
def AvgFilter(pre_avg, k, z):
    alpha = (k - 1) / k
    avg = alpha * pre_avg + (1 - alpha) * z
    return avg
```

```python
# 전압 측정값 생성 함수
def GetVolt():
    # 평균이 0, 표준편차가 4인 정규분포를 따르는 잡음 생성
    w = 0 + 4 * np.random.randn(1)
    # 측정 전압의 평균은 14.4V로 위의 잡음이 섞인 값을 측정값으로 설정
    z = 14.4 + w
    return z
```

```python
# 측정 간격 0.2 
dt = 0.2

# arange 함수를 이용해 0 ~ 10까지 0.2 간격으로 수 생성 
# 여기서는 10초까지의 측정값 개수를 구하기 위해 사용
t = np.arange(0, 10, dt)

# 측정값 개수
Nsamples = len(t)

# 측정값과 평균필터를 통과한 값을 저장할 array 생성 
Zsaved = np.zeros(Nsamples)
Avgsaved = np.zeros(Nsamples)

np.random.seed(1)
pre_avg = 0

# 새로운 측정값이 들어올 때마다 평균필터를 통과시켜 평균 계산 
for k in range(Nsamples):
    k += 1
    z = GetVolt()
    avg = AvgFilter(pre_avg, k, z)
    
    Zsaved[k-1] = z
    Avgsaved[k-1] = avg
    
    pre_avg = avg
    
Avgsaved
```

```python
plt.plot(t, Zsaved, 'r*--', label='Measured')
plt.plot(t, Avgsaved, 'o-', label='Average')
plt.legend(loc='lower left')
plt.title('Measured Voltages & Average Filter Values')
plt.xlabel('Time [sec]')
plt.ylabel('Volt [V]')
plt.show()
```
